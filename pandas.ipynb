```python
import pandas as pd
```

```python
flavors_df = pd.read_csv("datasets/Flavors.csv")
flavors_df
```

```python
# Groupby Base Flavor and get mean rating
base_flavor_groupby = flavors_df.groupby('Base Flavor')
base_flavor_groupby.mean()
```

```python
# Get count of rows in each group
base_flavor_groupby.count()
```

```python
# Perform multiple aggregations on single series
base_flavor_groupby.agg({'Total Rating': ['min','max','mean']})
```

```python
# Group by Base Flavor and Liked and get the mean ratings
flavors_df.groupby(['Base Flavor','Liked']).mean()
```
